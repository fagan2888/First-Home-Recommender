```python
def LinModel(X, y, estimator):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2)
    model = estimator
    model.fit(X_train, y_train)
    predicted = model.predict(X_test)
    print('{}: {}'.format(estimator, model.score(X, y)))
```
